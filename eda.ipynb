```python
# Baseline Correction & Normalization:
# Apply baseline correction to remove background noise.
# Normalize the spectral data for consistency across samples.

# Baseline correction
def baseline_correction(spectrum):
    baseline = (spectrum.iloc[0] + spectrum.iloc[-1]) / 2
    return spectrum - baseline

data = data.apply(baseline_correction, axis=1)
data
```

```python
print("Original DataFrame shape:", data.shape)
max_values = data.max(axis=1)
print("Max values Series shape:", max_values.shape)

data = data.div(max_values, axis=0)  # Normalize each spectrum by its maximum value
print("Normalized DataFrame shape:", data.shape)
```

```python
data
```

```python
# Plot the spectra
plt.figure(figsize=(10, 6))
for i in range(len(data)):
    plt.plot(data.iloc[i])
plt.xlabel('Wavelength')
plt.ylabel('Intensity')
plt.title('Spectral Intensity Across Wavelengths (after baseline correction)')
plt.legend()
plt.show()
```

```python
# Noise Reduction:
# Use smoothing techniques like Savitzky-Golay filter or moving average.

from scipy.signal import savgol_filter

data_smooth = data.apply(lambda x: savgol_filter(x, 21, 3), axis=1)
data_smooth
```

```python
type(data_smooth)
```

```python
# convert data_smooth to a DataFrame
data_smooth = pd.DataFrame(data_smooth.tolist(), index=data.index, columns=data.columns)
data_smooth
```

```python
# Plot the smoothed spectra
plt.figure(figsize=(10, 6))
for i in range(len(data_smooth)):
    plt.plot(data_smooth.iloc[i])
plt.xlabel('Wavelength')
plt.ylabel('Intensity')
plt.title('Smoothed Spectral Intensity Across Wavelengths')
plt.legend()
plt.show()
```

```python
# Multiple Scatter Correction:
# Implement scatter correction methods such as MSC or SNV to address scattering effects.

from sklearn.decomposition import PCA 
from sklearn.preprocessing import StandardScaler

def msc(data):
    # Apply mean centering
    data_mean_centered = data.sub(data.mean(axis=1), axis=0)
    
    # Perform PCA
    pca = PCA(n_components=3)
    scores = pca.fit_transform(data_mean_centered)
    
    # Reconstruct the data using the principal components
    data_corrected = pca.inverse_transform(scores)
    
    return data_corrected

data_msc = msc(data_smooth)
data_msc = pd.DataFrame(data_msc, columns=data.columns)
data_msc
```

```python
# Plot the MSC-corrected spectra
plt.figure(figsize=(10, 6))
for i in range(len(data_msc)):
    plt.plot(data_msc.iloc[i])
plt.xlabel('Wavelength')
plt.ylabel('Intensity')
plt.title('MSC-Corrected Spectral Intensity Across Wavelengths')
plt.legend()
plt.show()
```

# Regression Models

```python
# 3. Regression Models
# Steps:

# Baseline Regression Models:
# Fit Ridge regression, LASSO, and Elastic Net models to the labeled calibration data.
# Partial Least Squares (PLS) Regression:
# Use PLS to predict metabolite concentrations and assess variable importance.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(data_msc, target, test_size=0.2, random_state=42)
```

```python
# Fit Ridge regression model
ridge = Ridge(alpha=0.1)
ridge.fit(X_train, y_train)

# Predict the target variable
y_pred_ridge = ridge.predict(X_test)

# Evaluate the model
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

print("Ridge Regression:")
print("MSE:", mse_ridge)
print("R^2:", r2_ridge)
```

```python
# Fit LASSO regression model
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)

# Predict the target variable
y_pred_lasso = lasso.predict(X_test)

# Evaluate the model
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)

print("LASSO Regression:")
print("MSE:", mse_lasso)
print("R^2:", r2_lasso)
```

```python
# Fit Elastic Net model
elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic_net.fit(X_train, y_train)

# Predict the target variable
y_pred_elastic_net = elastic_net.predict(X_test)

# Evaluate the model
mse_elastic_net = mean_squared_error(y_test, y_pred_elastic_net)
r2_elastic_net = r2_score(y_test, y_pred_elastic_net)

print("Elastic Net Regression:")
print("MSE:", mse_elastic_net)
print("R^2:", r2_elastic_net)
```

```python
# Partial Least Squares (PLS) Regression:
# Fit PLS regression model to the labeled calibration data.
# Predict metabolite concentrations and assess variable importance.

from sklearn.cross_decomposition import PLSRegression

# Fit PLS regression model
pls = PLSRegression(n_components=3)
pls.fit(X_train, y_train)

# Predict the target variable
y_pred_pls = pls.predict(X_test)

# Evaluate the model
mse_pls = mean_squared_error(y_test, y_pred_pls)
r2_pls = r2_score(y_test, y_pred_pls)

print("PLS Regression:")
print("MSE:", mse_pls)
print("R^2:", r2_pls)
```

```python
# Variable Importance in PLS:
# Extract the variable importance in projection (VIP) scores from the PLS model.
# Identify the most important spectral features for predicting metabolite concentrations.

# Get the VIP scores
vip_scores = pls.x_scores_.var(axis=0) * pls.x_weights_ / pls.x_weights_.var(axis=0)
vip_scores = pd.Series(vip_scores.sum(axis=1), index=data_msc.columns)
```

```python
# Plot the VIP scores
plt.figure(figsize=(10, 6))
vip_scores.plot(kind='bar')
plt.xlabel('Wavelength')
plt.ylabel('VIP Score')
plt.title('Variable Importance in PLS')
plt.show()
```
