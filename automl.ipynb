```python
import numpy as np
import pandas as pd
import h2o
from h2o.automl import H2OAutoML
```

```python
data = pd.read_csv('data/data_w_target.csv')
```

```python
h2o.init()
h2o_data = h2o.H2OFrame(data)
```

```python
h2o_data
```

```python
aml = H2OAutoML(max_models=20, seed=1)
aml.train(x=data.columns.tolist()[:-6], y='Formate (g/L)', training_frame=h2o_data)
```

```python
lb = aml.leaderboard
print(lb)
```
